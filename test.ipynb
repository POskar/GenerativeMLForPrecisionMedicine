```python
import pandas as pd
df = pd.read_csv('data/HAD.csv')

array = []

for i in range(df.shape[0]):
    if df['MRS_90'][i] >= df['PREV_MRS'][i]:
        array.append(0)
    else:
        array.append(1)

df_comparison = df[['PREV_MRS', 'MRS_90']].copy()
df_comparison['MRS_COMPARISON'] = pd.DataFrame(array)
print(df_comparison.to_string())
```
